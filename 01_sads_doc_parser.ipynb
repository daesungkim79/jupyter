```python
import os
import itertools
from tqdm import trange
```

```python
def read_ptc_item(id):
    dict = {}
    os.system("im viewissue --user=p13870 --password=daesungkim {}>./ptcData/{}.txt 2>&1".format(id, id))

    with open("./ptcData/" + id + ".txt", "rt") as f:
        for line in f:
            try:
                key = line.split(":")[0].strip()
                val = line.split(":")[1].strip().splitlines()
            except Exception as e:
                pass
            dict.update({key:val})
            
    df = pd.DataFrame.from_dict(dict, orient='index', columns=["value"])            

    return df
```

```python
def find_swc_trace_info(id):
    trace_dict ={}
    
    df = read_ptc_item(id)
    
    try:
        up_trace_count = df.loc["Upstream Trace Count", "value"]
        down_trace_count = df.loc["Downstream Trace Count", "value"]
        implements_count = df.loc["Implements", "value"]
        decomposes_to_count = df.loc["Decomposes To", "value"]
        verified_by_count = df.loc["Verified By", "value"]
        sw_category = df.loc["SW Category", "value"]
        
        if implements_count == None: implements_count = 0
        else: implements_count = len(implements_count.split(','))
        
        if decomposes_to_count == None: decomposes_to_count = 0
        else: decomposes_to_count = len(decomposes_to_count.split(','))
            
        if verified_by_count == None: verified_by_count = 0
        else: verified_by_count = len(verified_by_count.split(','))            

        # 1 upstream
        upstream_id_1 = df.loc["Contained By", "value"]
        df_1 = read_ptc_item(upstream_id_1)

        # 2 upstream
        upstream_id_2 = df_1.loc["Contained By", "value"]
        df_2 = read_ptc_item(upstream_id_2)
        # find SWC name
        swcName = df_2.loc["Summary", "value"]

#         print("Id: {}".format(id))
#         print("Software Component Name: {}".format(swcName))
#         print("Upstream Trace Count: {}".format(up_trace_count))
#         print("Downstream Trace Count: {}".format(down_trace_count))
#         print("Implements Count: {}".format(implements_count))
#         print("Decompose To Count: {}".format(decomposes_to_count))
#         print("Verified By Count: {}".format(verified_by_count))
#         print("SW Category: {}".format(sw_category))

        trace_dict.update({swcName:[id, up_trace_count, down_trace_count, implements_count,\
                                   decomposes_to_count, verified_by_count, sw_category]})
    except Exception as e:
#        print(e)
        pass
        
    return trace_dict    
    
```

```python
import itertools
from tqdm.notebook import tqdm

def find_ptc_id(ptc_id):
    ls = ''
    
    df = read_ptc_item(ptc_id)

    try:
        ls = df.loc["Contains", "value"].split(",")

    except Exception as e:
        #print(e)
        pass

    content_id = []
    for i in ls:
        content_id.append(i.strip().split("ay")[0])
    
    return content_id

def det_depth(ptc_id, depth):
    ptc_id_temp = list()
    ptc_id_total = list()
    ptc_id_dict = dict()
    
    while depth != 0:
        if len(ptc_id) != 0:
            for i in tqdm(range(len(ptc_id)), desc='parsing id, detph-'+str(depth)):
                ptc_id_temp.append(find_ptc_id(ptc_id[i]))
        ptc_id_temp = list(itertools.chain(*ptc_id_temp))
        ptc_id = ptc_id_temp
        ptc_id_temp = list()
        ptc_id_total.extend(ptc_id)
        depth -= 1

    print(ptc_id_total)
    for r in tqdm(range(len(ptc_id_total)), desc='identifying category'):
        df_cat = read_ptc_item(ptc_id_total[r])
        cat = df_cat.loc["Category", "value"]
        ptc_id_dict.update({ptc_id_total[r]:cat})
    
    return ptc_id_dict
        
# depth 1
ptc_id_1 = find_ptc_id('2312420')
print(ptc_id_1)
# parse category (comment / heading / hw/sw speicification)
ptc_id_dict = det_depth(ptc_id_1, 3)
ptc_id_dict
```
