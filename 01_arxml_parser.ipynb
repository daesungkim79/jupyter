```python
import os

def searchFile(tarDir, extension, filenamearr, dirPath):
    dirArr = []
    if os.path.isdir(tarDir) == True:
        subDir = os.listdir(tarDir)
        for dirName in subDir:
            if 'src' in dirName or 'ar' in dirName:
                #print(dirName)
                dirArr.append(dirName)
            else:
                if os.path.isdir(tarDir+'/'+dirName) == True:
                    subDir_1 = os.listdir(tarDir+'/'+dirName)
                    for dirName_1 in subDir_1:
                        if 'src' in dirName_1 or 'ar' in dirName:
                            #print("dirName_1", dirName + '/' + dirName_1)
                            dirArr.append(dirName + '/' + dirName_1)

        for dir_line in dirArr:
            print(dir_line)
            full_filename = os.path.join(tarDir, dir_line)
            files = os.listdir(full_filename)
            #ext = os.path.splitext(full_filename)[-1]
            for file in files:
                ext = file.split('.')[-1]
                if ext == extension:
                    filenamearr.append(file)
                    dirPath.append(os.path.join(tarDir, file))
#              if os.path.isdir(full_filename):
#                 searchFile(full_filename, extension, filenamearr, dirPath)
                
    return filenamearr, dirPath
```
